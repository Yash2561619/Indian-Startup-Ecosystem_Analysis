```python
pf=pd.read_csv("indian_startup_funding_2020_2025_sample.csv")
```

```python
pf.head()
```

```python
pf.rename(columns={'InvestmentAmount_USD': 'InvestmentAmount_Rs'}, inplace=True)
```

```python
def in_ind(dollar):
    ind=dollar*90.72
    return ind/10000000
```

```python
pf['InvestmentAmount_Rs']=pf['InvestmentAmount_Rs'].apply(in_ind)
```

```python
pf.head()
```

```python
pf['Investors']=pf['Investors'].str.split(',').str.get(0)
```

```python
pf.head(10)
```

```python
pf.groupby('Investors')['Startup'].count().sort_values(ascending=False).head(10)
```

```python
pf.head()
```

```python
pf.groupby('InvestmentType')['Investors'].count().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%0.1f%%')
```

```python
pf['Date']=pd.to_datetime(pf['Date'],errors='coerce')
```

```python
pf.head()
```

```python
pf['Year']=pf['Date'].dt.year
```

```python
pf.head()
```

```python
pf[pf['Year']<2025].groupby('Year')['Startup'].count().sort_values(ascending=False).plot(kind='bar')
```

```python
year_counts = (
    pf[pf['Year'] < 2025]
    .groupby('Year')['Startup']
    .count()
    .sort_values(ascending=False)
)

colors = plt.cm.tab10(range(len(year_counts)))  # basic color palette

year_counts.plot(kind='bar', color=colors)

plt.xlabel('Year')
plt.ylabel('Number of Startups Funded')
plt.title('Startups Funded per Year')
plt.show()
```

```python
cm=pf.groupby('Startup')
cm['InvestmentAmount_Rs'].sum().sort_values(ascending=False).head(20)
```

```python
import matplotlib.pyplot as plt

top10 = (
    pf.groupby('Startup')['InvestmentAmount_Rs']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

colors = plt.cm.tab10(range(len(top10)))  # basic distinct colors

top10.plot(kind='bar', color=colors)

plt.xlabel('Startup')
plt.ylabel('Total Investment (INR)')
plt.title('Top 10 Startups by Funding')
plt.show()
```

```python
import matplotlib.pyplot as plt

top20 = (
    pf.groupby('Startup')['InvestmentAmount_Rs']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

ax = top20.plot(
    kind='bar',
    figsize=(12,6),
    colormap='viridis'   # color depends on investment value
)

plt.xlabel('Startup')
plt.ylabel('Total Investment (INR)')
plt.title('Top 20 Startups by Investment Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt

top20 = (
    pf.groupby('Startup')['InvestmentAmount_Rs']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

ax = top20.plot(
    kind='bar',
    figsize=(12,6),
    colormap='coolwarm'   # color depends on investment value
)

plt.xlabel('Startup')
plt.ylabel('Total Investment (INR)')
plt.title('Top 20 Startups by Investment Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```
